```python
import seaborn as sns
```

```python
## In the starting with Seaborn , first we will saw how to do Basic Plotting with seaborn 
## Here first function we are probabaly going to used  is (load_dataset)
#  Load_Dataset .. this is a Method in Seaborn,,, which will help us to Get some Free Dataset's...which are already There in seaborn ...
tips=sns.load_dataset('tips')
tips
# This Data have Categorical Data as well as , also have some Numerical Data   
```

```python
### Scatter Plot 
## As we are Creating Scatter Plot in Matplolib we can Also create Here 
import matplotlib.pyplot as plt

## Here I am trying to plot a Scatter-plot between a "Total bill" and "tip" 
sns.scatterplot(x='total_bill',y='tip',data=tips,color='blue') ##  # Here X- axis is total_bill ,, Y-axis is "tip" and "data" is my total entire Data i.e "tips"
plt.title(" scatter Plot of Total Bill Vs Tip ")

## The given Dataset is of an Restaurent ... if you  clearly Observes that 
## BLUE IS ALSO A DEFAULT COLOR ,, IF YOU NOT MENTIONED THE COLOR , BLUE WILL BE PRINTRD 
```

```python
## Line Plot 

sns.lineplot(x='size',y='total_bill',data=tips)###
plt.title(" This is a total bill Line Plot by  size ")
plt.show()
```

```python
#### There is one Fantastic thing about Matplotlib and Seaborn that Here you can Create Plots for Catgorical data ... there Are few example of Categorikal Data is :- 1) sex 2) Days 3) Time 
## Now , let's Understand how to plot them 

# Categorical Plots :- Here there are Diffreent -Diffrent plots we see All plots one - by one 
    # 1) Bar Plot 
sns.barplot(x='day',y='total_bill',data=tips) # Here I want to see total_bills per Day 
plt.title(" Bar Plot of Total bills per day ")
plt.show()
```

```python
    # 2) Box plot :- It is the Convient way to see Outliers in Data 

sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Total bill per day ")
```

```python
#### Now , in Corresponding to See the Distribution of the Data we will use "Violin plot"

sns.violinplot(x='day',y='total_bill',data=tips)
plt.title(" In-order to see the Distribution of Data")
```

```python
## Let's try some Histogram 
sns.histplot(tips['total_bill'],bins=10,kde=True)
# kde :- Is Kernel density is used to Put curver Like Line In Histogram
# Kernel Density = Is used to Give the shape to the Histogram.. so it Could better to understand the Comcepts 
```

```python
sns.histplot(tips['total_bill'],bins=10,kde=False)
# If I just Make it False then , shape will be removed 
```

```python
## When you want to Only see the Shape of your Graph/Plot , then we can also plot only KdE 
# kde Plot:- 
sns.kdeplot(tips['total_bill'],fill=True) # fill is new version word ,, shade was depricated Version
```

```python
## 3) Pairplot:- This is one of the Plotting technique which is used to ..." See Relationships between All the Varaibles and Difffrent -difrent Varaibles  "
sns.pairplot(tips)
## This Are Applicable for [" Numerical Varaible "]...Might not be Working for [" Catrgorical Data "]

### Here You Can see All the Relationships ....Between All the Varaibles and All diffrent-Diffrent Varaibles 
```

```python
## Heat-Map :- Is used to vizsaualize Co-relation between "[Numerical Varaible]"....it is also Might not Supported for "Catogorical Data"
corr=tips[['total_bill','tip','size']].corr()
corr
## You can see the Co-rrelation ship between [" Numerical Varaible"]
```

```python
### NOW HEAT-MAP WILL GIVE YOU THE NICE COMPARISON BETWEEN ALL THIS THREE AND ALSO PROVIDE " HIGH vIZVALZATION "
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
########## LET'S GO AHEAD AND SEE SOME PRACTICAL EXAMPLE #############
import pandas as pd
# Let's 1st read the csv file 
sales_df=pd.read_csv('data.csv')
sales_df.head()
```

```python
##### Let's Go ahead and Plot , Total Sales by product 

plt.figure(figsize=(10,7)) # Here I am Specifically using my figuare size is 10 inches wide  and 6 inches tall 
sns.barplot(x='Category',y="Sales",data=sales_df,estimator=sum)

## estimator :- Inside this you can write any Data  Aggregattion function which you wanted to used like sum,mean,median,max all this things ... estimator will find All those Values 
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x='Region',y="Sales",data=sales_df,estimator=sum)
plt.title('Total Sales by Region')
plt.xlabel('Product')
plt.ylabel("Total Sales")
plt.show()
```
