```python
### Heat-Map :- Specifically Use for Correlation , here I am try to figuare out some correlation between "Cars and its various Functions  "

sns.heatmap(mtcars.corr(),cbar=True,linewidths=0.5)
# cbar :- This Argument is used to draw the Color Bar 

## This error is Coming because , Somewhere our Data has "Non-numeric Values " and ... for the Smooth Excution of "Heat-Map" It wants only "NUmeric values "
```

```python
###### To fix this Error

# Load the mtcars dataset
mtcars = sns.load_dataset("mpg")  # Use an appropriate dataset 
# Ensure only numeric columns are used in the Dataset 
mtcars_numeric = mtcars.select_dtypes(include=['number'])

# Compute the correlation matrix
correlation_matrix = mtcars_numeric.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, cbar=True, annot=True, linewidths=0.5)
plt.title("Heatmap of Correlations")
plt.show()
```

```python
###### IN SUMMARY THE HEATMAP GET CORRELATION THAT
""" WHENEVER  OUR NO OF CYLINDERS INCRESES OUR MILES PER GALLON DECRESES """
# Not only this There is Co-relation of Every single "Category comparing with other "
```

```python
################### TO WORK WITH SOME SCATTER PLOT WE WANT SOME IRIS DATASET LET'S WORK WITH SOME IRIS DATASET ####################

# 1) Load_Datset :- 
iris=sns.load_dataset('iris')
iris.head()
```

```python
 # Let's print the count of each species
iris['species'].value_counts()
```

```python
### Now print the Scatter Plot for the folowing Species
sns.scatterplot(x='sepal_length',y='petal_length',data=iris)  
```

```python
## Now we can classify the diffrent Species of flower by "hue" parameter 
sns.scatterplot(x='sepal_length',y='petal_length',data=iris,hue='species')  
```

```python
#### CONCLUSION :- 1) SETOSA PLANT HAVE LOWEST SEPAL LENGTH AND PETAL LENGTH 
    #               2) VIRGINICA HAVE HIGHEST SEPAL LENGTH AND PETAL LENGTH 
```

```python
## Now let's try to Plot :- Pair-Plot 
res=sns.pairplot(iris)
```

```python
### DIAGONAL PLOTS ARE "HISTOGRAM" AND REAMING ARE SCATTER PLOTS 
```

```python
# Now let's try to Vizualize Some gruops of Data together , by using ("hue") this Parameter 
# hue= This Paramter is used to Vizualaize Some Grops/Sets of Data Toghether  
sns.pairplot(iris,hue="species",palette="Set1")
```

```python
## FROM ALL OF THE GRAPHS SHOWING HERE , THE "PETAL LENGTH " HAVE VERY NICE CLASSIFICATION ....
```

```python
## LMPLOT :- IS USED FOR REGRESSION LINE AND MOST REGRESSION GRAPHS 
res= sns.lmplot(x='petal_length',y='petal_width',data=iris)
# The line you see here is " Linear Refression line "
```

```python
#### HERE YOU  CAN SET "HUE" PARAMETER FOR "FOR MORE CLASSIFICATION AND CLARIFICATION BETWEEN GROUPS AND  SETS"
### ALSO YOU CAN USE MARKERS FOR KNOWING SOME "INSIGHTS FULL FROM THIS "

#let's see how to do that 
```

```python
res=sns.lmplot(x='petal_length',y='petal_width',hue="species",data=iris,markers=["o",'*',"^"])

# HERE IN MARKERS I HAVE USED THREE DIFFRENT MARKERS FOR , THREE DIFFRENT SPECIES OF PLANTS / FLOWERS WHERE:
```
