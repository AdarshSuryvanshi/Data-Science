```python
df_fillNa=df.fillna(0)
df # Now if you could see All After this All misssing values are replaced with zero
```

```python
df['Sales_fillNa']=df['Sales'].fillna(df['Sales'].mean())
df  # All NaN(Missing Values) are replace with mean values (Sales) this cloumn becoz.... there are 3 Missing values are there in (Sales) in this cloumn ..we need to remove them   
```

```python
df.dtypes
```

```python
## Rename Columns
df=df.rename(columns={'Sales_Fillna':'Sales_fillNA'}) #Sales_Fillna
df.head()
```

```python
## Change Datatypes 
df['Sale_New']=df['Sales'].fillna(df['Sales'].mean()).astype(int) ## (Sales_new= This  cloumn contains (int) type of Sales )
df.head()
```

```python
df['New sale Value']=df['Sale_New'].apply(lambda x:x*2)
df.head()
```

```python
## Single Gruoping cloumn

grouped_mean=df.groupby('Product')['Sale_New'].mean()
print("This is the mean ",grouped_mean)   
# Here in this I am going to find out average sale of Each Product 
```

```python
print('\n')
grouped_sum=df.groupby('Product')['Sales'].sum()
print(" This is the sum",grouped_sum)   
# Here in this I am going to find out Total  sale of Each Product 
```

```python
print('\n')
grouped_max=df.groupby('Product')['Sales'].max()
print(" this is the max",grouped_max)   
# Here in this I am going to find out max  sale of Each Product
```

```python
# Mulitiple Grouping cloumn
print('\n')
grouped_max=df.groupby(["Product","Region"])['Sales'].max()
print("This is the max of ",grouped_max) # Here in this I am going to find out max  sale of Each Product According to Diffrent Regions  
 
```

```python
grouped_mean=df.groupby(["Product","Region"])['Sales'].mean()
grouped_mean  
# Here in this I am going to find out average sale of Each Product According to Diffrent Regions 
```

```python
print('\n')
grouped_sum=df.groupby(["Product","Region"])['Sales'].sum()
grouped_sum  
# Here in this I am going to find out Total  sale of Each Product According to Diffrent Regions  
```

```python
## aggregate multiple functions
groudped_agg=df.groupby('Region')['Sales'].agg(['mean','sum','count'])
groudped_agg
```
