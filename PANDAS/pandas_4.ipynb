```python
#### SOME EXAMPLES WRT TO PANDAS 
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# ---------------------------
# 1. Creating a DataFrame
# ---------------------------
# Sample data for demonstration
data = {
    "Product": ["A", "B", "A", "C", "B", "A"],
    "Region": ["North", "South", "East", "North", "East", "South"],
    "Sales": [100, 200, 150, 300, 400, 250],
    "Quantity": [5, 10, 7, 8, 12, 6]
}

# Creating a DataFrame
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# ---------------------------
# 2. Reading and Writing CSV
# ---------------------------
# Saving DataFrame to a CSV file
df.to_csv("sales_data.csv", index=False)
print("\nDataFrame saved as 'sales_data.csv'.")

# Reading the same CSV file
df_csv = pd.read_csv("sales_data.csv")
print("\nDataFrame read from 'sales_data.csv':")
print(df_csv)

# ---------------------------
# 3. Inspecting Data
# ---------------------------
print("\nInspecting DataFrame:")
print(df.info())  # Information about the DataFrame
print("\nSummary Statistics:")
print(df.describe())  # Statistical summary of numerical columns

# ---------------------------
# 4. Selecting Data
# ---------------------------
# Selecting a single column
print("\nSales Column:")
print(df["Sales"])

# Selecting multiple columns
print("\nSales and Quantity Columns:")
print(df[["Sales", "Quantity"]])

# Selecting rows using conditions
print("\nRows where Sales > 150:")
print(df[df["Sales"] > 150])

# ---------------------------
# 5. Adding and Modifying Columns
# ---------------------------
# Adding a new column for Total Revenue
df["Revenue"] = df["Sales"] * df["Quantity"]
print("\nDataFrame after adding Revenue column:")
print(df)

# Creating a new column based on conditions
df["Performance"] = df["Sales"].apply(lambda x: "High" if x > 200 else "Low")
print("\nDataFrame after adding Performance column:")
print(df)

# ---------------------------
# 6. Handling Missing Data
# ---------------------------
# Introducing missing values for demonstration
df.loc[1, "Sales"] = None
print("\nDataFrame with missing values:")
print(df)

# Filling missing values with the mean of the column
df["Sales"] = df["Sales"].fillna(df["Sales"].mean())
print("\nDataFrame after filling missing values:")
print(df)

# ---------------------------
# 7. Grouping and Aggregation
# ---------------------------
# Grouping by Product and calculating total Sales
grouped = df.groupby("Product")["Sales"].sum()
print("\nTotal Sales by Product:")
print(grouped)

# ---------------------------
# 8. Pivot Table
# ---------------------------
# Creating a pivot table for average sales
pivot = df.pivot_table(values="Sales", index="Product", columns="Region", aggfunc="mean")
print("\nPivot Table (Average Sales):")
print(pivot)

# ---------------------------
# 9. Visualizing Data
# ---------------------------
# Plotting sales distribution
print("\nPlotting Sales Distribution...")
df["Sales"].plot(kind="hist", title="Sales Distribution", color='skyblue')
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()

# ---------------------------
# 10. Handling Time-Series Data
# ---------------------------
# Adding a Date column for demonstration
df["Date"] = pd.date_range(start="2023-01-01", periods=len(df), freq="D")
print("\nDataFrame with Date column:")
print(df)

# Setting Date as index and resampling by week
df.set_index("Date", inplace=True)
weekly_sales = df["Sales"].resample("W").sum()
print("\nWeekly Sales (Resampled Data):")
print(weekly_sales)

# ---------------------------
# 11. Merging and Joining
# ---------------------------
# Creating another DataFrame for merging
additional_data = {
    "Product": ["A", "B", "C"],
    "Category": ["Electronics", "Furniture", "Clothing"]
}
df_additional = pd.DataFrame(additional_data)

# Merging DataFrames on Product
merged_df = pd.merge(df.reset_index(), df_additional, on="Product")
print("\nMerged DataFrame:")
print(merged_df)

# ---------------------------
# 12. Saving Final DataFrame
# ---------------------------
merged_df.to_csv("final_data.csv", index=False)
print("\nFinal DataFrame saved as 'final_data.csv'.")
```
