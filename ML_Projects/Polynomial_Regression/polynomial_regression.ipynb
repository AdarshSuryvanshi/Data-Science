```python
## Here we are LEARNING  How  To Create OUr Own  DATASET By using Randome Library   


X=6*np.random.rand(100,1)-3
Y=0.5 *X**2 +1.5*X+2 +np.random.randn(100,1)
# Quadratic Equation used :- y=0.5x^2+1.5x+2+outliers
plt.scatter(x=X,y=Y,color="g")
plt.xlabel('X dataset')
plt.ylabel(" Y dataset ")
```

```python
#### HERE YOU CAN SEE A POLYNOMIAL REGRESSION LINE IS GETTING CREATED APPRANTLY HERE 

# First we will try to Implement using "Simple Linear Regression "
# Then will go and try to implement using "Polynomial Linear Regression "
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
# Lets implement Simple Linear Regression 
from sklearn.linear_model import LinearRegression
regression_1=LinearRegression()
```

```python
regression_1.fit(X_train,Y_train)
```

```python
from sklearn.metrics import r2_score
score=r2_score(Y_test,regression_1.predict(X_test))
print(score)

# HERE YOU CAN SEE ACCURACY IS VERY LESS (62.00)% ...BEUSE WE ARE USING 
""" 
"SIMPLE LINEAR REGRESSION "...BECAUSE OF USING THIS ACCURACY IS VERY LESS, ON WE WILL SEE IN VISUALIZATION 
"""
```

```python
###Lets Visualize this 
plt.plot(X_train,regression_1.predict(X_train),color="r")
plt.scatter(X_train,Y_train)
plt.xlabel("X Dataset")
plt.ylabel(" Y Dataset")
```

```python
#### SO TO INCREASE THE ACCURACY WE WILL TRY TO USE "POLYNOMIAL REGRESSION " SO OUR LINE WILL TRY TO PASS WITH MAXIMUM POINTS AND MAKE A SMOOTH CURVE SO ACCURACY WILL INCEARSE 
# SO LET'S APPLY POLYNOMIAL TRANSFORMATION 
# Degree =1 , i.e Simple Linear Regression 
# Degree =2 , i.e Multiple(Polynomial Regression )
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2,include_bias=True)
X_train_ploy=poly.fit_transform(X_train)
X_test_ploy=poly.fit_transform(X_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression

# Initialize and fit the linear regression model
regression_2 = LinearRegression()
regression_2.fit(X_train_ploy, Y_train)

# Use the predict method to get predictions
Y_pred = regression_2.predict(X_test_ploy)

# Calculate the R2 score
score_2 = r2_score(Y_test, Y_pred)
print(score_2)
```

### Pipeline in Polynimial :- 

```python
### What is A Pipelineling in Polynomial :- 
""" 
In this WE have to "Create " a Generic Fuctions , by using  that Generic Functions whichever Degree You will Assign , That Kind of Curve you will see Over there ... Along with some Predictions of 
New Data Sets  
"""
```

```python
"""from sklearn.pipeline import Pipeline
def Poly_regression(degree):
    X_new =np.linspace(-3 ,3,200).reshape(200,1)

    poly_features=PolynomialFeatures(degree=degree,include_bias=True)
    lin_reg=LinearRegression([
        ("Poly_features",poly_features),
        ("lin_reg",lin_reg)
    ])
    Poly_regression.fit(X-X_train,Y_train)
    Y_pred_new=Poly_regression.predict(X_new)
    # Plotting Predicions Line 
    plt.plot(X_new ,Y_pred_new,"r",label="Degree"+str(degree),linewidth=3)
    plt.plot(X_train ,Y_train ,"b.",linewidth=3)
    plt.plot(X_test  ,Y_test,"g.",linewidth=3)
    plt.legend(loc="Upper Left")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.axis([-4,4,0,10])
    plt.show()
"""
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline

def Poly_regression(degree, X_train, Y_train, X_test, Y_test):
    X_new = np.linspace(-3, 3, 200).reshape(200, 1)

    # Creating a pipeline for polynomial regression
    poly_regression = Pipeline([
        ("Poly_features", PolynomialFeatures(degree=degree, include_bias=True)),
        ("lin_reg", LinearRegression())
    ])

    # Fitting the model
    poly_regression.fit(X_train, Y_train)

    # Making predictions
    Y_pred_new = poly_regression.predict(X_new)

    # Plotting the prediction line
    plt.plot(X_new, Y_pred_new, "r", label="Degree " + str(degree), linewidth=3)
    plt.plot(X_train, Y_train, "b.", linewidth=3)
    plt.plot(X_test, Y_test, "g.", linewidth=3)
    
    plt.legend(loc="upper left")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.axis([-4, 4, 0, 10])
    plt.show()
```

```python
## Now You Can Simply Pass , The Degree in the Fuction and You will get the "Best fit Line According to That "
 # For Degree =1
Poly_regression(1, X_train, Y_train, X_test, Y_test)
### WE get Simple Linear Regression 
```

```python
# For Degree =2
Poly_regression(2, X_train, Y_train, X_test, Y_test)
```

```python
# For Degree=3 
Poly_regression(3, X_train, Y_train, X_test, Y_test)
```
