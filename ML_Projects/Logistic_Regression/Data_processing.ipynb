```python
### In this  Implementation , we will be creating Our own Dataset and then we will be creating a Model on that Dataset .

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification  # It is having its own Dataset. By this we can create our own Dataset.
# The Benefit of doing this is :- This Dataset is Already Standardized , we don't have to do Standardiztion  and Normalization on this Dataset 
# We can directly  use this Dataset for Training our Model , we can Direclty Focus on Implementaion Part 
```

```python
## Let's Create the Dataset
X ,Y = make_classification(n_samples=1000 ,n_features=10 , n_classes=2 ,random_state= 42)
 # Here You are seeing this X and Y , this are My Independent and Dependent Features Respectively..
 # n_samples=1000 :- Means I am Creating 1000 rows
 # n_features=10 :- Means I am creating 10 columns 
 # n_Classes =2 :- I am Creating 2 Classes 
  # randome State :- It is the Seed Value so  that we can get the same value as we run the code Everytime
    
```

```python
pd.DataFrame(X) # This is My Independent Features 
pd.DataFrame(Y) # This is my Dependent Features 
# Like that Here , we have Succesfully Created our Own Dataset, also we have Succesfully Craeted Dependent / Independent Features 
```

```python
pd.DataFrame(X)
```

```python
Y
```

```python
### Now , Let's Split the Dataset into "Train-Test"
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y ,test_size=0.25 ,random_state=42)
```

```python
#Model Training 
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression() # Here Inside this There is one Parameter Called as "Calss_weight -" is Imp in Imbalaced Dataset , It will tell you to Focus on which  Dataset or Data  
```

```python
logistic.fit(X_train,Y_train) # Model Training 
Y_pred=logistic.predict(X_test)
print(Y_pred)
```

```python
logistic.predict_proba(X_test) # It will give you the Probability of the Prediction 
```

```python
logistic.score(X_test,Y_test) # It will give you the Accuracy of the Model
```

```python
# Let's Check the Accuarcy using Performance Matrix 
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
score =accuracy_score(Y_test,Y_pred)
print(score)
cm= confusion_matrix(Y_test,Y_pred)
print(cm)

print(classification_report(Y_test,Y_pred))

## Here Accuracy I'm getting is 84%...
```

```python
  # What is Hyper -paramter tunning ? Why we do this ?
 # --> Hyper-Parameter Tunning is the Process of Choosing the Optimal Parameter for the Model , so that we can get the Best Model
# We can do this becoz , we can get the Best Model and we can Get the Best Accuracy of the Model 
```

### HYPER-PARAMTER TUNNING AND CROSS VALIDATION

```python
model =LogisticRegression() # 
penalty =["l1","l2","elasticnet"] # It is the Regularization Parameter , It is used to avoid Over -fitting 
C_values=[100,10,1.0,0.1,0.01] # C_Values it is also --//-- 
solver=["newton-cg","lbfgs","liblinear","sag", "saga"] # It is the Optimization Algorithm , which is used to find the Optimal Parameter
```

```python
params=dict(penalty=penalty, C=C_values,solver=solver) # It is the Dictionary , which is having the Key and Value Pair 
params
```

```python
###Grid Search CV 
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
Cv=StratifiedKFold()
grid = GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=Cv,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train,Y_train)
grid.best_params_
```

```python
###Here you can See 
""" 
{'C': 0.01, 'penalty': 'l2', 'solver': 'newton-cg'}

This are the Most Imporatant Features which are selected by the Grid After doing the Hyperparamter Tunning 
This Featurse Are selected by the Grid 
"""
```

```python
grid.best_score_
### This is the Best Accuarcy We got Till now 87% 
```
