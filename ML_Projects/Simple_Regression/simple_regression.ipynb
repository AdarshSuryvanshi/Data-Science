```python
## Now We will Diffrentiate and Break- down the Data-set into 2 Parts
#  "Independent Features" and "Dependent Features "
df.head()
```

```python
X=df[["Weight"]] ## Independent Feature
# Independent Features are Always... / Schould be in the form of either "Data-Frame" or "2-D Array "..
# [[]] :- Always  "Independent - Features" Is form mai Hona Chahiye
print(X)
type(X) ## This is in the form of Data-frame .. and it schould be 
```

```python
## It is Fine To Stored "Dependent Features" In a 1-D Array , But "Independent Features" are Must Store in a "2-D Array" or "Data-Frame "
Y= df["Weight"]
print(Y)
type(Y)
```

```python
X_seies= df['Weight']
## This is usefull when you are Understanding the concept of "LinearRegression"
# Just remember this is "1-D array "
```

TILL HERE WE HAVE SUCCESFULLY SEPERATED OUR "INDEPENDENT FEATURES" & "DEPENDENT FEATURES"

```python
#### LET'S SPLIT OUR DATA INTO "TRAINING" AND "TESTING"
from sklearn.model_selection import train_test_split
X_train , X_text ,y_train ,y_test = train_test_split(X,Y,test_size=0.25,random_state=42)

"""
Test Size: Specifies the proportion of the dataset to allocate for testing, ensuring the model 
is evaluated on unseen data. For example, test_size=0.2 reserves 20% of the data for testing and 80% for training.

Random State :- Ensures the data is split the same way every time you run the code.
 It helps get consistent results when training and testing your model.
 """
```

```python
# Now we are Doing "Standardization"

### Standardization :- The Reason for using Standardization we Already Discussed in Statistics Please go through that ...that Z-score vala 
""" 
We perform standardization to remove the impact of different 
units of measurement across features (e.g., meters, kilograms, seconds). 
This ensures all features are on the same scale, allowing the model to treat them equally without bias toward larger units.

"""
```

```python
X.shape , X_train.shape ,X_text.shape,Y.shape,y_test.shape,y_train.shape
# X_train is the 75% of the Whole Data and X_test is the 25% 
```

```python
### Prepossing is the Librarry of "Sk-Learn" which is used for "Standardization"
from sklearn.preprocessing import StandardScaler ## Standardization
```

```python
### Here I am Training My data
## Standardizing the Data 
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)

## Linear Regression is the "Distance-Based" and Optimization based Algorithum so we , We will be USing "Standardization " Here 
## By using this "fit_transform" My Whole "Independent Feature" this Column is get Converted into "Standardization"
## You can see All points are get "Standardized" in (Mean=0 and S.D=1)
## All are gets Standardized into 0 and 1
```

```python
X_train
```

```python
# Here I am Testing My Data 
X_text=scaler.transform(X_text)

## NOTE:- For Training The Data We used "fit_transform()" and For Testining the Data We used "Transform()"
# Becoz there is a Concept Called Data -Leackage 
```

```python
X_text
```

```python
"""
fit_transform() :- is used for "Training The Data " the Data
transform() :- Is used for "Testing the Data "  

"""
# MAKE SURE YOU DATA SCHOULD NOT BE LEEK FROM ANY WHERE , So that's why Becoz of the Concept Data leakage ,"fit_tansform and transform() are "useful 
```

```python
plt.scatter(X_train,y_train)
```

```python
## Apply  Simple Linear Regression 
from sklearn.linear_model import LinearRegression
# Here we Import the "LinearRegression" for Applying Algorithum  
```

```python
regression=LinearRegression() ## Regression Call over Here 
```

```python
# As we learn Above... 1-D Array is not Accepted in "Independent Features ", Now you will Understand Why ??? 
regression.fit(X_seies,Y) # Here I am Passing 1-D Array , See the error
""" Expected a 2-dimensional container but given 1-D array"""
 # That Means ML does not want 1-D array as a "Independent Fearures "...
 ## Output as "Dependent Feature "in 1- D Array will fine , 
```

```python
### So We will Provide "2-D Array "
regression=LinearRegression(n_jobs=-1)
regression.fit(X_train,y_train)

""" n_jobs :- in LinearRegression controls parallel processing. Setting it to -1 uses all available CPU cores to speed up computation.Help to Process Everything parally 

fit() :- in LinearRegression is used to train the model by learning patterns from the training data (features and labels). Fit() is used to Train the Model
"""
# In- Summary :- 1) fit_transform() :- Is used to "Trainning the Dataset"
# 2) transform() :- Is used to "Testining the Dataset"
# 3) fit():- Is used to "Trainning the Model"


""" Since we have  we have set above Our Training Data is 75% and Testing data 25 %"""
# Thaths why My (X_train and Y_train) both have 75% of Total data , as they are Responsible for Training the Model 
```

```python
regression ## Here my regression Showing This 
```

```python
### LET'S DO SOME AMAZING WORK 
## TRY TO FIND OUT THE "Coefficient = Slope"... Coefficient is nothing but my Slope
print("Coefficients or Slope is ",regression.coef_)
print("Intercepts is  is ",regression.intercept_)
```

```python
""" Here  it has Only 1 Coefficient that means , It is a "Simple Linear regression """
# Because It has only 1 Coefficient and 1 Slope 
# 1 Coefficient = 1 Input Column = 1 Independent Feature= Simple Linear Regression 
```

```python
#### Let's Visualize this  "Simple regression Line " and "Slope and Intercepts "
# Plot the Training Data  and plot the Best fit Line.... Here we are Plotting the "Best the Fit " for 
plt.scatter(X_train,y_train)
plt.plot(X_train,regression.predict(X_train))

## Here we are "Traning the Data " and creating the Best fit line ...
# And this Best fit Line Having (Coefficient = 16.220779489) and (Intercept =75.05888)
```

```python
""" What we are Doing ????
:- WE HAVE TRAIN OUR MODEL ON "TRAING DATA" AND NOW BASED ON THAT WE ARE GOING TO PERFOME OUR PREDICTIONS ON "TEST DATA "
"""
# In this Whole process the Data  of both  "Model and Predictions" are need to be Seprated...
```

### Prediction of Test Data

```python
#### Prediction for the Test Data """"
## Here we are Going to Predict our "Height" is our Dependent Varaible ON "Weight" is our Dependent Varaible 

Y_Prediction=regression.predict(X_text)
Y_Prediction,y_test


##
""" Since we have  we have set above Our Training Data is 75% and Testing data 25 %"""
# Thaths why My (X_test) have 25% as Test Data   , as on them  Prediction is happened see the Prediction of Inputs from (X_Test )
# See the Answer below this are the Predictions of X_test  25% Test Data  

```

```python
  ## let's Compare the Predictions of (X_Test) with  Actual Answers (Y_test) by using Performance Matrix 
from sklearn.metrics import mean_absolute_error,mean_squared_error 
```

```python
mse=mean_squared_error(y_test,Y_Prediction)
mae=mean_absolute_error(y_test,Y_Prediction)
rmse=np.sqrt(mse)

print(mse)
print(mae)
print(rmse)
```

```python
#### For R^2
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_test,Y_Prediction)
print(score)
```

```python
## Adjusted R^2
1-(1-score)*(len(y_test)-1)/(len(y_test)-X_text.shape[1]-1)
```
