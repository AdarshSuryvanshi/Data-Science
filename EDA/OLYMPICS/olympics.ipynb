```python
### Q) Print the Columns Name Containing Null Values or Missing Values in the form of a List 
def columns_with_nulls(athletics_df):
    # Initialize an empty list to store column names with nulls
    Column_name_list = []

    # Iterate through each column in the dataframe
    for column in  athletics_df.columns:
        # Check if the column contains any null or missing values
        if athletics_df[column].any():
            # Add the column name to the list
            Column_name_list.append(column)
        return Column_name_list
       

    # Print the list of column names with null values
print(columns_with_nulls(athletics_df))

## So this is the Perfect Code to findthe Columns Name Which having Null Values 

```

```python
#### Let's try to Convert some "Categorical Data" into "Numeric Data"
### 1st We will Try to Convert Sex Column , We can Separate (Male and Feamale).. So Here for Now we have Only 2 category.. which are (Male and Female) so We can "Use" onehotCodder
from sklearn.preprocessing import OneHotEncoder
```

```python
##create an instance of Onehotencoder
encoder=OneHotEncoder()
```

```python
## perform fit and transform
encoded=encoder.fit_transform(athletics_df[['Sex']]).toarray()
## Give your Data-Frame Name and Column Name On - Which you have to Apply "Onehotencoding"..it will give "Code" to every Unique Categorical Value and Convert them into "Numeric"
```

```python
### Here I am Creating an Another Data-Frame for "One-hot-encodder"
encoder_dataframe=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

```python
encoder_dataframe  ## Here I have Assigned a Code for Male and Female... for Male:- 1.0 and for Female :- 0.0
```

```python
### Now it's Time ConcAt this Data-Frame into the "Original " Dataframe 
# Merge this in the Original Dataframe 
pd.concat([athletics_df,encoder_dataframe],axis=1)
athletics_df.head()
```

```python
athletics_df.info()
```

```python
###Now we will Give Numeric Values To The "Season" Column 
## perform fit and transform
encoded_2=encoder.fit_transform(athletics_df[['Season']]).toarray()
## Give your Data-Frame Name and Column Name On - Which you have to Apply "Onehotencoding"..it will give "Code" to every Unique Categorical Value and Convert them into "Numeric"
```

```python
encoder_df_2=pd.DataFrame(encoded_2,columns=encoder.get_feature_names_out())
# Here It will give Unique Code To Unique Categorical Values in Season....
# Since we have Only Summer / Winter we have use "One-hot-Encodding"
```

```python
encoder_df_2
```

```python
### Now it's Time ConcAt this Data-Frame into the "Original " Dataframe 
# Merge this in the Original Dataframe 
pd.concat([athletics_df,encoder_df_2],axis=1)
athletics_df.head()
```
