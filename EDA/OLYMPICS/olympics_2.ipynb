```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load Dataset
athletics = pd.read_csv("athlete_events.csv")
region=pd.read_csv("noc_regions.csv")
```

```python
## Join the Data-Frame 
# Here will Join both the Datasets Together  
athletics_df=athletics.merge(region,how="left",on="NOC") # Here I have Merge my (region) Data-Frame into the atheletics Data-Frame On the Left side ...
# That Means My athletics Section will Go and Merge on the Left hand Side ... and region will stay on the Left hand side ....
athletics_df.head()
```

```python
athletics_df.info()
```

```python
#### Now let's Handled All this Missing Values and try to Stored Answer of This Missing Values in Copied Data-Frame 
athletics_df_copy=athletics_df.copy()
```

```python
athletics_df_copy.head()
```

```python
## Always remember Before Starting Handling Missing Values of Any Column .... Check that Column Belongs to Which Distribution , Accordingly you can 
## Use Techniques to solve that Missing Values based on type which you have Studied , for that "Missing Values "
```

```python
sns.histplot(data=athletics_df_copy["Age"],edgecolor='black',color='orange',kde=True)
```

Conclusion :- The above Distribution is "Left Skewed" .. So, The above Distribution is Follows "Log-Normal" Type Distributions ...we know that
            For Log-Normal Distributions we used "Median imputation Technique" to Handle Missing Values ..

```python
## Here try to Stored Solution in Another column
athletics_df_copy['Age_median']=athletics_df_copy['Age'].fillna(athletics_df_copy['Age'].median()) 
```

```python
athletics_df_copy[["Age","Age_median"]]
athletics_df_copy.info()
athletics_df_copy.isnull().sum()
```

```python
### Let's Check the Distribution for Height and Weight 
sns.histplot(data=athletics_df_copy["Height"],edgecolor="black",color='orange',kde=True)
## We can see it's Almost around Normally/Gaussian Distribution so we can Apply "Mean"..imputation Technique Now..
```

```python
athletics_df_copy['height_mean'] =athletics_df_copy["Height"].fillna(athletics_df_copy["Height"].mean()) 
```

```python
athletics_df_copy[["Height","height_mean"]]

### LIKE THIS WE CAN ALSO DO FOR "WEIGHT" AND OTHER ALL COLUMNS , WHO HAVE MISSING VALUES AND BASED ON THEIR DISTRIBUTIONS WE WILL MAKE OTHER SOLUTION COLUMNS..for Missing Values
```

```python
###  Let's Take only the Gold Medal Atheletes from the Dataset 
Gold_medal_athletes = athletics_df_copy[(athletics_df_copy.Medal == "Gold")]
Gold_medal_athletes.head()

## By using this We understood How to print Data of Specific Category from Diffrent Columns 
```

```python
## Take only the Values that are Diffrent from NaN...i.e Other from Missing Values ..All 
Gold_medal_athletes=Gold_medal_athletes[np.isfinite(Gold_medal_athletes['Age'])]
Gold_medal_athletes
```

```python
# Now we will The Athletes Who have Won "Gold - Medal" After the Age of 60
Gold_medal_athletes["ID"][Gold_medal_athletes["Age"] > 60].value_counts()
```

```python
## If we want Total Count 
Gold_medal_athletes["ID"][Gold_medal_athletes["Age"] > 60].count()
# There are total  6 Athletes who have Scored a Gold Medal After the Age of 60
```

```python
## Now we will Print in which Event/sports , they all Got 6 Medals
sporting_event =Gold_medal_athletes['Sport'][Gold_medal_athletes['Age'] > 60]
sporting_event
## In this Sports Atheletes got Gold AFter the age of 60

## In this way we can Seprate the Data From "Column"
```

```python
## Plot the Sporting Event ...
# Sporting Event is in (Dtype:- Object) i.e It is of "Categorical VAlues"...it is having Categorical Values That's why we are Printing Countplot
# Set the style for the plot
# Ensure sporting_event is a DataFrame
if isinstance(sporting_event, pd.Series):
    sporting_event = sporting_event.to_frame(name="Sport")

## Here We Convert "sporting_event" this Column in to the Data-Frame ........
# Beacuse the (data="") this parameter in Countplot Only Accepts (["DataFrame"]).... (data=) This parameter Only Accepts Data-Frame... Not Accept Any column or Something like that ..

sns.set(style="darkgrid")

# Create a figure with specified size
plt.figure(figsize=(10, 5))

# Create a countplot
sns.countplot(x="Sport", data = sporting_event, palette="Set1")

# Adjust layout to ensure everything fits
plt.tight_layout()

# Add a title for better understanding
plt.title("Count of Sporting Events by Sport", fontsize=16)

# Add labels for axes
plt.xlabel("Sport", fontsize=14)
plt.ylabel("Count", fontsize=14)

# Show the plot
plt.show()
```

```python
### Gold Medals from Each Country 

Gold_medal_athletes.region.value_counts().reset_index(name="Medal").head(10)
```

```python
# Ensure that 'index' is explicitly named
TotalGoldMedals = Gold_medal_athletes.region.value_counts().reset_index(name="Medal")

# Plot the data
g = sns.catplot(
    x="region", y="Medal", data=TotalGoldMedals.head(6),
    height=5, kind="bar", palette="rocket"
)
g.despine(left=True)
g.set_xlabels("Top 5 Countries")
g.set_ylabels("Number of Medals")
plt.title("Gold Medals per Country")
plt.show()
```
