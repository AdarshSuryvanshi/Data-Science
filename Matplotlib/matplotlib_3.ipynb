```python
## Create a 6*6 2-D Array let 1 and 0 be placed Alternatively across the Diagonals 
Z=np.zeros((6,6),dtype=int ) ## dtype= Data Type
Z
#### So we can Placed  1 and 0 alternatively by using Slicing 
Z[1::2,::2]=1
### Explanation:-
""" Here in  Sliced [1::2] :- Means Starting from row which having index=1,, then Directly jump to row=3,,, No need to go on Index=2.... Because is 2
    Here in  Sliced [::2] :- Means Starting from Column which having index=0,, then Directly jump to row=2,,, No need to go on Index=2.... Because is 1

    This is how Slicing works 
"""

Z
```

```python
# Find the Total number and Locations of Missing values in the Array 
Z=np.random.rand(10,10)
Z[np.random.randint(10,size=5),np.random.randint(10,size=5)]=np.nan
Z


## In 2-D Numpy-Array (nan) is Assigned to show the Missing values ... Whereas in 
## In Pandas Dataframe (Nan) is Assigned to show the Missing Values 
```

```python
print(" The total number of Missing Values are ;-\n",np.isnan(Z).sum())
print(" Indexexe  of Missing Values are ;-\n",np.argwhere(np.isnan(Z)))
inds=np.where(np.isnan(Z))
print(inds)
Z[inds]=0
Z

## 1st we Find out Where are the missing Valuers are there ,then after that ... We replaced that Missing Values with Zeros
```

```python
""" Here Is a Example of one Important libraary in Python use to generates Randome  Values and Numbers which are Helping to practising 
"""
import random

# Random float between 0 and 1
print("Random float (0 to 1):", random.random())

# Random integer between 10 and 20
print("Random integer (10 to 20):", random.randint(10, 20))

# Random float between 2.5 and 5.5
print("Random float (2.5 to 5.5):", random.uniform(2.5, 5.5))

# Random choice from a list
colors = ['red', 'blue', 'green']
print("Random color:", random.choice(colors))

# Random sample of 2 elements
print("Random sample of colors:", random.sample(colors, 2))

# Shuffle a list
deck = [1, 2, 3, 4, 5]
random.shuffle(deck)
print("Shuffled deck:", deck)

# Using a seed for reproducibility
random.seed(42)
print("Seeded random integer:", random.randint(1, 10))
```
