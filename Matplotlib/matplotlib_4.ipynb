```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,25)
y=x*x+2
print(x)
print(y)
print(np.array([x,y]).reshape(25,2).reshape(2,25))
```

```python
## It TAKES 1 COMMAND TO DRAW a line 
plt.plot(x,y,'r') # 'r'  stands for red
```

```python
# DRAW A SUBGRAPH 
plt.subplot(1,2,1) # Here, The contents of the Brackets representts the (rowa,column,indiexes).....NOTE:- Always In a Figure for multiple Diagrams "Rows" and "Columns" Schould be same...and indexes schould be Updated Repetadly ....after doing one diag..
plt.plot(x,y,'r--') # The third parametr here determines the color and Line style 
plt.show()

plt.subplot(1,2,2)
plt.plot(y,x,'g--')
```

```python
######### LET'S UNDERSTAND SOME NEW THINGS 
fig=plt.figure()
axis=fig.add_axes([0.5,0.1,0.8,0.8]) # Contril the Left , right , width , height of the canvas (from 0 to 1)
axis.plot(x,y,'r')
```

```python
# agin we can draw some subgraphs 
fig,axes=plt.subplots(nrows=1,ncols=2)  # Sub-Graphs is of 1 row, 2 columns 
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)  # Sub-Graphs is of 1 row, 2 columns 
for ax in axes:
    ax.plot(x,y,'r')
```

```python
### IN MATPLOTLIB YOU CAN SET OTHER PROPERTIES SUCH AS line color , transparency and more , 

fig,axes=plt.subplots(dpi=150) # It controls the size and fig  of the Graph 
axes.plot(x,x+1,color='red')
axes.plot(x,x+2,color='g')
axes.plot(x,x+3,color='y')
```

```python
## HOW TO SEE THE TRANSPARENCY..BY ALPHA 
fig,axes=plt.subplots(dpi=150)  
axes.plot(x,x+1,color='red',alpha=.5)
axes.plot(x,x+2,color='g',alpha=.5)
axes.plot(x,x+3,color='y',alpha=.5)

fig,axes=plt.subplots(dpi=150)  
axes.plot(x,x**2,color='red',alpha=.9)
axes.plot(x,x+2,color='g',alpha=.9)
axes.plot(x,x+3,color='y',alpha=.9)

fig,axes=plt.subplots(dpi=150)  
axes.plot(x,x**2,color='red',alpha=.3)
axes.plot(x,x+2,color='g',alpha=.5)
axes.plot(x,x+3,color='y',alpha=.5)
```

```python
### SET THE CANVAS GRID AND AXIS CHANGE 

fig,Axes=plt.subplots(1,2,figsize=(10,5))  #### HERE I AM TAKING 3 PARAMETERS SO USED " subplots".......while "subplot" is for 2 PARAMETR  
x=np.linspace(0,10,25)
Axes[0].plot(x,x**2,x,x**3,lw=2)
Axes[0].grid(True)

Axes[1].plot(x,x**2,x,x**3,lw=2)
Axes[1].grid(True)
Axes[1].set_ylim([0,60])
Axes[1].set_xlim([2,5])
```

```python
#### DRAW A HISTOGRAM 

n=np.random.randn(1000)
fig,axes=plt.subplots(1,2,figsize=(12,4))
axes[0].set_title("Deafult Histogram ")
axes[0].hist(n,edgecolor='black',color='orange')

axes[1].set_title("Cumulative  detailed  Histogram ")
axes[1].hist(n,bins=50,cumulative=True,edgecolor='black',color='orange')

### IF YOU DO NOT DID "CUMULATIVE=TRUE", THEN IT WILL BECOME A NORMAL HISTOGRAM 
```
