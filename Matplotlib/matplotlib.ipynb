```python
import matplotlib.pyplot as plt
import numpy as np

# ==========================
# 1. Line Plot Example
# ==========================
# Data for line plot
years = np.arange(2010, 2021)
sales = [15, 18, 20, 25, 30, 34, 40, 48, 50, 60, 70]

plt.figure(figsize=(10, 7)) # This figaure is Defining a Size of Complete White box , inside which we are going to uploding Images

plt.subplot(2, 3, 1)  # Add subplot (2 rows, 3 columns, 1st plot).... This Diagram will go and Placed on 1st plot 
plt.plot(years, sales, marker='o', color='blue', linestyle='-', label='Sales Trend')
plt.title("Line Plot - Sales Over Years")
plt.xlabel("Years")
plt.ylabel("Sales (in $k)")
plt.grid(True)
plt.legend()

# ==========================
# 2. Scatter Plot Example
# ==========================
# Data for scatter plot
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.rand(50) * 100  # Random sizes for points

plt.subplot(2, 3, 2)  # Add subplot (2nd plot).... This Diagram will go and Placed on 2nd plot 
plt.scatter(x, y, s=sizes, c=y, cmap='viridis', alpha=0.7, edgecolor='black')
plt.title("Scatter Plot - Random Data")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.colorbar(label="Color Intensity")

# ==========================
# 3. Bar Chart Example
# ==========================
# Data for bar chart
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 30]

plt.subplot(2, 3, 3)  # Add subplot (3rd plot)...... This Diagram will go and Placed on 3rd  plot 
plt.bar(categories, values, color='orange', edgecolor='black')
plt.title("Bar Chart - Category Values")
plt.xlabel("Categories")
plt.ylabel("Values")

# ==========================
# 4. Histogram Example
# ==========================
# Data for histogram
data = np.random.normal(0, 1, 1000)  # Generate random data

plt.subplot(2, 3, 4)  # Add subplot (4th plot).... This Diagram will go and Placed on 4th plot 
plt.hist(data, bins=20, color='purple', edgecolor='black', alpha=0.7)
plt.title("Histogram - Data Distribution")
plt.xlabel("Value Range")
plt.ylabel("Frequency")

# ==========================
# 5. Pie Chart Example
# ==========================
# Data for pie chart
segments = [30, 20, 25, 25]
labels = ['Segment A', 'Segment B', 'Segment C', 'Segment D']
explode=(0.2,0,0,0)

plt.subplot(2, 3, 5)  # Add subplot (5th plot)...... This Diagram will go and Placed on 5th plot 
plt.pie(segments, labels=labels, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightblue', 'lightgreen', 'pink'],explode=explode,shadow=True)
plt.title("Pie Chart - Market Segments")

# ==========================
# Finalize and Display
# ==========================
plt.tight_layout()
plt.show()
```
